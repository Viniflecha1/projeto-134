```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/content/PRO-Stars-Dataset-CSVs/star_with_gravity.csv")

df.head()
```

```python
#filtrando as estrelas com base na distância
bools =[]
for d in df.Distance:
    if d<=100:
        bools.append(True)
    else:
        bools.append(False)
```

```python
is_dist = pd.Series(bools)
```

```python
is_dist.head()
```

```python
star_dist=df[is_dist]
```

```python

star_dist.reset_index(inplace=True,drop=True)
star_dist.head()
```

```python
star_dist.shape
```

```python
#mantenha as estrelas com gravidade semelhante à do Soll
gravity_bool = []
for g in star_dist.Gravity:
    if g<=350 and g>=150:
        gravity_bool.append(True)
    else :
        gravity_bool.append(False)
```

```python
is_gravity = pd.Series(gravity_bool)
```

```python
final_stars = star_dist[is_gravity]
final_stars.head()
```

```python
final_stars.shape
```

```python
final_stars.reset_index(inplace=True,drop=True)
```

```python
final_stars.head()
```

```python
final_stars.to_csv("filtered_stars.csv")
```
